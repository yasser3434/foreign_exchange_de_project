### Fetch historical data 

```python
import os
import requests
import pandas as pd
from datetime import datetime, timedelta
from dotenv import load_dotenv
```

```python
load_dotenv()

api_key = os.getenv("api_key")
# api_key
```

```python
base_currency = "EUR"
target_currencies = ["NOK", "EUR", "SEK", "PLN", "RON", "DKK", "CZK"]
```

```python
start_date = datetime(2026, 2, 26)
end_date = datetime.today()
```

```python
records = []
current_date = start_date
```

```python
current_date
```

```python
start_date
```

```python
while current_date < end_date:
    year    = current_date.year
    month   = current_date.month
    day     = current_date.day

    url = f"https://v6.exchangerate-api.com/v6/{api_key}/history/EUR/{year}/{month}/{day}"
    response = requests.get(url)
    data = response.json()

    try:
        rates = data["conversion_rates"]
        record = {"date" : current_date.strftime("%Y-%m-%d")}
        
        for currency in target_currencies:
            record[currency] = rates.get(currency) 
        
        records.append(record)

        current_date += timedelta(days=1)
    except Exception as e:
        print("ERROR : ", e) 
```

```python
records
```

```python
df = pd.DataFrame(records)
df
```

```python
cross_pairs = []
```

```python
for _, row in df.iterrows():
    for base, target in permutations(currencies, 2):
        rate = row[target] / row[base]
        cross_pairs.append({
            "date": row["date"],
            "base_currency": base,
            "target_currency": target,
            "rate": round(rate, 6)
        })
```

## Dimensional table : dim currency 

```python
currencies
```

```python
dim = []

for currency in currencies:
    response = requests.get(f"https://v6.exchangerate-api.com/v6/{api_key}/enriched/EUR/{currency}")
    data = response.json()

    if data.get("result") == "success":
        dim.append({
            "locale" : data["target_data"]["locale"]
            , "two_letter_code" : data["target_data"]["two_letter_code"]
            , "currency_name" : data["target_data"]["currency_name"]
            , "currency_name_short" : data["target_data"]["currency_name_short"]
        })
    else:
        print(data.get("result"))
```

#### Date dim

```python
from datetime import datetime, timedelta
```

```python
date_records = []
current = datetime(2025, 1, 1)
end = datetime.today()
```

```python
cursor.execute("""
    CREATE TABLE IF NOT EXISTS dim_date(
        date TEXT PRIMARY KEY
        , year INTEGER NOT NULL
        , month INTEGER NOT NULL
        , day INTEGER NOT NULL
        , quarter INTEGER NOT NULL
    );
""")
```

```python
while current < end:
    date_records.append((
        current
        , current.year
        , current.month
        , current.day
        , (current.month - 1) // 3 + 1
    ))

    current += timedelta(days=1)
```

```python
date_records[:3]
```

```python
cursor.executemany("""
    INSERT INTO dim_date(date, year, month, day, quarter)
    VALUES(?, ?, ?, ?, ?)
""", date_records)
```
